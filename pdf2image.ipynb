```python
from pdf2image import convert_from_path
from dotenv import load_dotenv
import os
import easyocr
import numpy as np
```

```python
load_dotenv() # Load the variables

poppler_path = os.getenv("poppler_path")
```

```python
reader = easyocr.Reader(['en'], gpu=False)
```

```python
def convert_pdf(pdf_path, save_dir, res=400):
    pages = convert_from_path(pdf_path,res, poppler_path=poppler_path)

    name_with_extension = pdf_path.rsplit("/")[-1]
    name = name_with_extension.rsplit('.')[0]

    for idx, page in enumerate(pages,start=1):
        page.save(f'{save_dir}/{name}_{idx}.png', 'PNG')

    for idx, page in enumerate(pages,start=1):
        print(f"\n--- Page {idx} ---")
        result = reader.readtext(np.array(page), detail=0, paragraph=True)
        print("\n".join(result))


convert_pdf("/Users/LENOVO/Desktop/OCR project/thumb_1200_1699.pdf","/Users/LENOVO/Desktop/OCR project/")
```
