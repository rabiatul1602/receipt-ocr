```python
import pytesseract
from PIL import Image
```

```python
img = Image.open("D:\LENOVO\Downloads\WhatsApp Image 2026-06-20 at 00.59.23.jpeg")
```

```python
pytesseract.pytesseract.tesseract_cmd = r'D:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DATAFRAME)
data = data.dropna(subset=["text"])
data["text"] = data["text"].astype(str).str.strip()
data = data[data["text"] != ""]
```

```python
keyword = "Credit"
matches = data[data["text"].str.contains(keyword, na=False)]

for _, match in matches.iterrows():
    line_num = match["line_num"]
    block_num = match["block_num"]
    par_num = match["par_num"]

    same_line = data[
        (data["block_num"] == block_num) &
        (data["par_num"] == par_num) &
        (data["line_num"] == line_num)
    ].sort_values("word_num")

    line_text = " ".join(same_line["text"].tolist())
    print(line_text)
```

```python
import pytesseract
from PIL import Image
import re

#img = Image.open('rot_-4.png')  # use the deskewed version
data = pytesseract.image_to_data(img, config='--psm 6', output_type=pytesseract.Output.DATAFRAME)
data = data.dropna(subset=['text'])
data['text'] = data['text'].astype(str).str.strip()
data = data[data['text'] != '']

# Find any line containing a money-amount pattern (XX.XX)
money_pattern = re.compile(r'^\d{1,5}[.,]\d{2}$')

for (block, par, line), group in data.groupby(['block_num','par_num','line_num']):
    texts = group.sort_values('left')['text'].tolist()
    for t in texts:
        if money_pattern.match(t.replace(',', '')):
            print('Line:', texts)
```
